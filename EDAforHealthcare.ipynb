**General Thumb Rules:**

- For features with less missing values - can use regression to predict the missing values or fill with the mean of the values present, depending on the feature.
- For features with very high number of missing values - it is better to drop those columns as they give very less insight on analysis.
- As there's no thumb rule on what criteria do we delete the columns with high number of missing values, but generally you can delete the columns, if you have more than 30-40% of missing values.

```python
def uniplot(df,col,title,hue = None):

    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30


    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)

    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright')

    plt.show()
```

```python
uniplot(new_df1_target1,col='Hypertension',title='Distribution of Gender for NoShow Customers',hue='Gender')
```

```python
uniplot(new_df1_target0,col='Hypertension',title='Distribution of Gender for NoShow Customers',hue='Gender')
```

#### 1. **Gender Distribution**

- No-show rates are slightly higher among females, but the difference is not significantly disproportionate.

#### 2. **Age Group Trends**
- The age groups **0â€“20** show higher attendance (~80%) compared to older age bands.
- Most age groups exhibit a fairly balanced show-to-no-show ratio, though younger patients (especially infants) tend to adhere to appointments more reliably.

#### 3. **Scholarship Participation**
- ~99,666 patients didnâ€™t receive a scholarship, with ~80% of them showing up.
- Among the 21,801 patients who received scholarships, ~75% attended â€” suggesting scholarship availability may not strongly correlate with higher attendance.

#### 4. **Hypertension Status**
- Patients with hypertension (21,801) had an impressive ~85% show rate.
- Those without hypertension had a lower show rate (~78%), indicating patients with chronic conditions may be more committed to scheduled care.

#### 5. **Diabetes Status**
- Diabetic patients (7,943) showed up ~83% of the time.
- Non-diabetic individuals had a slightly lower attendance (~80%), consistent with the pattern seen in hypertensive patients.

#### 6. **SMS Notifications**
- Surprisingly, those **who received SMS reminders** had a lower show rate (~72%).
- This may suggest either ineffective timing of notifications or potential overreliance on messaging without follow-up.

#### 7. **Appointment Day Trends**
- Appointments are **rare on Saturdays** and **nonexistent on Sundays**, indicating scheduling is typically constrained to weekdays.
- Highest appointment volumes and attendance occur between Monday and Friday.

#### 8. **Neighborhood Influence** *(Before Dropping)*
- Neighborhood-level attendance rates clustered around 80%, with minimal variance.
